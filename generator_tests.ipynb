```python
from llm_project.generator import ResponseGenerator
from bert_score import score as bert_score
import logging
import pandas as pd
logging.getLogger("transformers").setLevel(logging.ERROR)

HF_TOKEN = ''
```

```python
def calculate_bertscore(response, etalon):
    P, R, F1 = bert_score(
        [response],
        [etalon],
        lang="ru",
        verbose=False,
        model_type="bert-base-multilingual-cased"
    )
    return F1.item()
```

```python
def compare_answers(question, ref_answer):
    agent_answer = resp_gen.generate(question)
    agent_answer = agent_answer.rsplit('\n\n', 1)[0]
    print('Ответ агента: ', agent_answer)
    print('Эталонный ответ: ', ref_answer)
    score = calculate_bertscore(agent_answer, ref_answer)
    print('bert score = ', round(score, 3))
    print('-' * 50)
```

```python
resp_gen = ResponseGenerator(hf_token = HF_TOKEN)
```

# Проверка на содержательное соответствие кодексам

```python
df = pd.read_csv('questions_answers.csv')
df
```

```python
for (x, y) in zip(df['question'].values[:10], df['reference_answer'].values[:10]):
    compare_answers(x, y)
```
